# Annual Investment Schedule Backtests (USA)

```python
import pandas as pd

# Set the plotting DPI settings to be a bit higher.
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [7.0, 4.5]
plt.rcParams['figure.dpi'] = 150
```

```python
r=pd.read_csv('backtestBenchmarkRuns_Random_Annual.csv', index_col=0)
# see the average relative return, and the number of backtests
print(r.iloc[-1].mean(), r.iloc[-1].count())
```

```python
neuralNet=pd.read_csv('backtestBenchmarkRuns_neuralNet_Annual.csv', index_col=0)
print(neuralNet.iloc[-1].mean(), neuralNet.iloc[-1].count())
```

```python
randomForest=pd.read_csv('backtestBenchmarkRuns_randomForest_Annual.csv', index_col=0)
print(randomForest.iloc[-1].mean(), randomForest.iloc[-1].count())
```

```python
extraTrees=pd.read_csv('backtestBenchmarkRuns_extraTrees_Annual.csv', index_col=0)
print(extraTrees.iloc[-1].mean(), extraTrees.iloc[-1].count())
```

```python
gradBoost=pd.read_csv('backtestBenchmarkRuns_gradBoost_Annual.csv', index_col=0)
print(gradBoost.iloc[-1].mean(), gradBoost.iloc[-1].count())
```

```python
knn=pd.read_csv('backtestBenchmarkRuns_KNN_Annual.csv', index_col=0)
print(knn.iloc[-1].mean(), knn.iloc[-1].count())
```

```python
pd.DataFrame({'random':r.iloc[-1],
              'neuralNet':neuralNet.iloc[-1],
              'randomForest':randomForest.iloc[-1],
              'extraTrees':extraTrees.iloc[-1],
              'gradBoost':gradBoost.iloc[-1],
              'knn':knn.iloc[-1]
              }).boxplot()
plt.ylim([0, 8])
plt.title('Annual Selection US Stocks',fontsize=16);

plt.ylabel('Relative Return');
```
